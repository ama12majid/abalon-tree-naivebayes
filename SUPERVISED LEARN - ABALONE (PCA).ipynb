```r
# memanggil package party (untuk model decision tree) dan package e1071 (untuk naive bayes classifier)
library(party)
library(e1071)
```

```r
#pembentukkan environment data
#pembagian data menjadi learning set dan testing set
#kami mengambil n-fold menjadi 5
#sample: 836 836 835 835 835

#learning dan testing1
n1 <- length(abalone[,1])
case.sample1 <- sample(1:n1, 836)
abalone.test1 <- abalone[case.sample1, ]
abalone.learn1 <- abalone[-case.sample1, ]

#learning dan testing2
n2 <- length(abalone.learn1[,1])
case.sample2<- sample(1:n2, 836)
abalone.test2 <- abalone.learn1[case.sample2, ]
sisa2 <- abalone.learn1[-case.sample2, ]
abalone.learn2 <- rbind(sisa2, abalone.test1)

#learning dan testing3
n3 <- length(sisa2[,1])
case.sample3<- sample(1:n3, 835)
abalone.test3 <- sisa2[case.sample3, ]
sisa3 <- sisa2[-case.sample3, ]
abalone.learn3 <- rbind(sisa3, abalone.test1, abalone.test2)

#learning dan testing4
n4 <- length(sisa3[,1])
case.sample4<- sample(1:n4, 835)
abalone.test4 <- sisa3[case.sample4, ]
sisa4 <- sisa3[-case.sample4, ]
abalone.learn4 <- rbind(sisa4, abalone.test1, abalone.test2, abalone.test3)

#learning dan testing5
abalone.test5 <- sisa4
abalone.learn5 <- rbind(abalone.test1, abalone.test2, abalone.test3, abalone.test3)
```

```r
#memodelkan atribut-atribut terhadap atribut Rings.kategori dengan model tree

abalone.tree1 <- ctree(Rings.kategori ~ Sex + Length, data = abalone.learn1)
abalone.tree2 <- ctree(Rings.kategori ~ Sex + Length, data = abalone.learn2)
abalone.tree3 <- ctree(Rings.kategori ~ Sex + Length, data = abalone.learn3)
abalone.tree4 <- ctree(Rings.kategori ~ Sex + Length, data = abalone.learn4)
abalone.tree5 <- ctree(Rings.kategori ~ Sex + Length, data = abalone.learn5)
# menampilkan model tree yang terbentuk
plot(abalone.tree1)
plot(abalone.tree2)
plot(abalone.tree3)
plot(abalone.tree4)
plot(abalone.tree5)
```

```r
#prediksi model tree terhadap testing model tree

abalone.prediksi1 <- predict(abalone.tree1, abalone.test1)
abalone.prediksi2 <- predict(abalone.tree2, abalone.test2)
abalone.prediksi3 <- predict(abalone.tree3, abalone.test3)
abalone.prediksi4 <- predict(abalone.tree4, abalone.test4)
abalone.prediksi5 <- predict(abalone.tree5, abalone.test5)
```

```r
#membuat tabel cm untuk prediksi dan aktual dari data testing model tree

abalone.CM1 <- table(abalone.test1[,4], abalone.prediksi1)
abalone.CM2 <- table(abalone.test2[,4], abalone.prediksi2)
abalone.CM3 <- table(abalone.test3[,4], abalone.prediksi3)
abalone.CM4 <- table(abalone.test4[,4], abalone.prediksi4)
abalone.CM5 <- table(abalone.test5[,4], abalone.prediksi5)

#tampilkan tabel CM

abalone.CM1
abalone.CM2
abalone.CM3
abalone.CM4
abalone.CM5
```

```r
#ukuran akurasi model tree
akurasi1 <- (sum(diag(abalone.CM1)))/sum(abalone.CM1)
akurasi2 <- (sum(diag(abalone.CM2)))/sum(abalone.CM2)
akurasi3 <- (sum(diag(abalone.CM3)))/sum(abalone.CM3)
akurasi4 <- (sum(diag(abalone.CM4)))/sum(abalone.CM4)
akurasi5 <- (sum(diag(abalone.CM5)))/sum(abalone.CM5)

akurasi.tree <- data.frame(akurasi1, akurasi2, akurasi3, akurasi4, akurasi5)
akurasi.tree
mean.akurasi.tree <- (akurasi1+akurasi2+akurasi3+akurasi4+akurasi5)/5
mean.akurasi.tree 
```

```r
#NAIVE BAYES
```

```r
#memodelkan atribut-atribut terhadap atribut Rings dengan algoritma naive bayes

nb1 <- naiveBayes(Rings.kategori ~ Sex + Length, data = abalone.learn1)
nb2 <- naiveBayes(Rings.kategori ~ Sex + Length, data = abalone.learn2)
nb3 <- naiveBayes(Rings.kategori ~ Sex + Length, data = abalone.learn3)
nb4 <- naiveBayes(Rings.kategori ~ Sex + Length, data = abalone.learn4)
nb5 <- naiveBayes(Rings.kategori ~ Sex + Length, data = abalone.learn5)
```

```r
#prediksi model tree terhadap testing naive bayes

prediksi.nb1 <- predict(nb1, abalone.test1)
prediksi.nb2 <- predict(nb2, abalone.test2)
prediksi.nb3 <- predict(nb3, abalone.test3)
prediksi.nb4 <- predict(nb4, abalone.test4)
prediksi.nb5 <- predict(nb5, abalone.test5)
```

```r
#membuat tabel cm untuk prediksi dan aktual dari data testing naive bayes
cm.nb1 <- table(abalone.test1[,4], prediksi.nb1)
cm.nb2 <- table(abalone.test2[,4], prediksi.nb2)
cm.nb3 <- table(abalone.test3[,4], prediksi.nb3)
cm.nb4 <- table(abalone.test4[,4], prediksi.nb4)
cm.nb5 <- table(abalone.test5[,4], prediksi.nb5)

#tampilkan tabel cm bayes
cm.nb1
cm.nb2
cm.nb3
cm.nb4
cm.nb5
```

```r
#ukuran akurasi naive bayes
akurasi.1 <- (sum(diag(cm.nb1)))/sum(cm.nb1)
akurasi.2 <- (sum(diag(cm.nb2)))/sum(cm.nb2)
akurasi.3 <- (sum(diag(cm.nb3)))/sum(cm.nb3)
akurasi.4 <- (sum(diag(cm.nb4)))/sum(cm.nb4)
akurasi.5 <- (sum(diag(cm.nb5)))/sum(cm.nb5)

akurasi.nb <- data.frame(akurasi.1, akurasi.2, akurasi.3, akurasi.4, akurasi.5)
akurasi.nb
mean.akurasi.nb <- (akurasi.1+akurasi.2+akurasi.3+akurasi.4+akurasi.5)/5
mean.akurasi.nb 
```

```r
#true positive model tree
TP.tree1 <- abalone.CM1[1,1]
TP.tree2 <- abalone.CM2[1,1]
TP.tree3 <- abalone.CM3[1,1]
TP.tree4 <- abalone.CM4[1,1]
TP.tree5 <- abalone.CM5[1,1]
TP.tree <- c(TP.tree1, TP.tree2, TP.tree3, TP.tree4, TP.tree5)

#true negative model tree
TN.tree1 <- abalone.CM1[2,2]
TN.tree2 <- abalone.CM2[2,2]
TN.tree3 <- abalone.CM3[2,2]
TN.tree4 <- abalone.CM4[2,2]
TN.tree5 <- abalone.CM5[2,2]
TN.tree <- c(TN.tree1, TN.tree2, TN.tree3, TN.tree4, TN.tree5)

#false positive model tree
FP.tree1 <- abalone.CM1[2,1]
FP.tree2 <- abalone.CM2[2,1]
FP.tree3 <- abalone.CM3[2,1]
FP.tree4 <- abalone.CM4[2,1]
FP.tree5 <- abalone.CM5[2,1]
FP.tree <- c(FP.tree1, FP.tree2, FP.tree3, FP.tree4, FP.tree5)

#false negatif model tree
FN.tree1 <- abalone.CM1[1,2]
FN.tree2 <- abalone.CM2[1,2]
FN.tree3 <- abalone.CM3[1,2]
FN.tree4 <- abalone.CM4[1,2]
FN.tree5 <- abalone.CM5[1,2]
FN.tree <- c(FN.tree1, FN.tree2, FN.tree3, FN.tree4, FN.tree5)


presisi.tree <- TP.tree/(TP.tree+FP.tree)
recall.tree <- TP.tree/(TP.tree+FN.tree)
presisi.tree
recall.tree

#pengukuran True Positive Rate dan False Positife Rate model tree
TPR.tree <- TP.tree/(TP.tree+FN.tree)
FPR.tree <- FP.tree/(TN.tree+FP.tree)
cbind(TPR.tree, FPR.tree)
```

```r
#true positive naive bayes
TP.nb1 <- cm.nb1[1,1]
TP.nb2 <- cm.nb2[1,1]
TP.nb3 <- cm.nb3[1,1]
TP.nb4 <- cm.nb4[1,1]
TP.nb5 <- cm.nb5[1,1]
TP.nb <- c(TP.nb1, TP.nb2, TP.nb3, TP.nb4, TP.nb5)

#true negative naive bayes
TN.nb1 <- cm.nb1[2,2]
TN.nb2 <- cm.nb2[2,2]
TN.nb3 <- cm.nb3[2,2]
TN.nb4 <- cm.nb4[2,2]
TN.nb5 <- cm.nb5[2,2]
TN.nb <- c(TN.nb1, TN.nb2, TN.nb3, TN.nb4, TN.nb5)

#false positive naive bayes
FP.nb1 <- cm.nb1[2,1]
FP.nb2 <- cm.nb2[2,1]
FP.nb3 <- cm.nb3[2,1]
FP.nb4 <- cm.nb4[2,1]
FP.nb5 <- cm.nb5[2,1]
FP.nb <- c(FP.nb1, FP.nb2, FP.nb3, FP.nb4, FP.nb5)

#false negatif naive bayes
FN.nb1 <- cm.nb1[1,2]
FN.nb2 <- cm.nb2[1,2]
FN.nb3 <- cm.nb3[1,2]
FN.nb4 <- cm.nb4[1,2]
FN.nb5 <- cm.nb5[1,2]
FN.nb <- c(FN.nb1, FN.nb2, FN.nb3, FN.nb4, FN.nb5)

#pengukuran presisi dan recall naive bayes

presisi.nb <- TP.nb/(TP.nb+FP.nb)
recall.nb <- TP.nb/(TP.nb+FN.nb)
presisi.nb
recall.nb

#pengukuran True Positive Rate dan False Positife Rate naive bayes
TPR.nb <- TP.nb/(TP.nb+FN.nb)
FPR.nb <- FP.nb/(TN.nb+FP.nb)
roc.nb <- cbind(TPR.nb, FPR.nb)
names(roc.nb) <-c("TPR.nb", "FPR.nb")
roc.nb
```

```r
eval.roc <-  cbind(TPR.tree, FPR.tree, TPR.nb, FPR.nb)
eval.roc
```
